```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3,4])
y = np.array([500,600,700,800])

print(f"Size : {x}")
print(f"Price : {y}")
```

```python
plt.scatter(x,y)
plt.xlabel("Size of Houses")
plt.ylabel("Price of Houses")
plt.title("House Price prediction")
plt.show()
```

```python

def predf(x,w,b):

  m = x.shape[0]
  pred= np.zeros(m)

  for i in range(m):
      pred[i]=x[i]*w + b

  return pred
```

```python
pred= predf(x,100, 20)

plt.scatter(x,y)
plt.plot(x, pred)
```

# Cost Functon

```python
#print(x,y)

w=210
b=90

pred= predf(x,w,b)


def cost_fun(y, pred):

  m= y.shape[0]
  sq=0

  for i in range(m):
     sq= sq+ (pred[i]-y[i])**2

  return ( 1/(2*m) )*sq


print(cost_fun(y,pred))
```
